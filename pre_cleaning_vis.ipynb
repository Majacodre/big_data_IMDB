```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("data/merged_train_imdb_data.csv")
validation_data = pd.read_csv("data/merged_validation_hidden_imdb_data.csv")
test_hidden_data = pd.read_csv("data/merged_hidden_imdb_data.csv")
```

## Train data statistics

```python
train_summary = pd.DataFrame({
    "count": train_data.count(),
    "nulls": train_data.isnull().sum(),
    "null_%": train_data.isnull().mean() * 100,
    "unique": train_data.nunique()
})

train_summary
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
train_summary.sort_values("null_%", ascending=False).plot(
    y="null_%",
    kind="bar",
    ax=axes[0]
)
axes[0].set_title("Missing Values")

train_summary.sort_values("unique", ascending=False).plot(
    y="unique",
    kind="bar",
    ax=axes[1]
)
axes[1].set_title("Unique Values Count")

fig.suptitle("Train data statistics by Column (Before Cleaning)")

plt.show()
```
